```python
def znajdz_centyl(wiek_dni, pomiar, tabela):
    # Znajdź wiersz w tabeli najbliższy podanemu wiekowi
    najblizszy_wiek = tabela.iloc[(tabela['Day'] - wiek_dni).abs().argsort()[:1]]

    # Iteruj przez kolumny centyli (od P01 do P999)
    for centyl in ['P01', 'P1', 'P3', 'P5', 'P10', 'P15', 'P25', 'P50', 'P75', 'P85', 'P90', 'P95', 'P97', 'P99', 'P999']:
        if pomiar <= najblizszy_wiek[centyl].iloc[0]:
            return centyl  # Zwróć nazwę kolumny, która reprezentuje centyl

    return "Pomiar poza zakresem"  # Jeśli pomiar jest poza zakresem tabeli

# Użycie funkcji
wiek_dni_dziecka = 365 * 2  # Przykładowy wiek w dniach (np. 2 lata)
pomiar_dziecka = 85  # Przykładowy pomiar (np. wzrost w cm)
centyl = znajdz_centyl(wiek_dni_dziecka, pomiar_dziecka, df_wybrany)
print(f"Dziecko znajduje się w {centyl} centylu")
```

```python
import matplotlib.pyplot as plt

# Zdefiniowanie zakresu wiekowego
min_wiek = 0       # minimalny wiek w dniach (od urodzenia)
max_wiek = 1825    # maksymalny wiek w dniach (5 lat)

# Przygotowanie danych
# Wybierz zakres wiekowy i odpowiednie kolumny (na przykład dla wzrostu)
df_wybrany = df_wybrany[(df_wybrany['Day'] >= min_wiek) & (df_wybrany['Day'] <= max_wiek)]

# Tworzenie wykresu
plt.figure(figsize=(10, 6))
for centyl in ['SD4neg', 'SD3neg', 'SD2neg', 'SD1neg', 'SD0', 'SD1', 'SD2', 'SD3', 'SD4']:
    plt.plot(df_wybrany['Day'], df_wybrany[centyl], label=f'{centyl}')

plt.xlabel('Wiek w dniach')
plt.ylabel('Pomiar (np. wzrost w cm)')
plt.title('Siatka Centylowa dla Chłopców')
plt.legend()
plt.grid(True)
plt.show()

```
